```python
# Define function to plot 
def megahit_comp_plot(data, param, outfp):
    plt.figure() # to prevent seaborn from plotting multiple plots onto 1 
    plot = sns.barplot(x='sampleID', y=param, hue='Input', data=data).get_figure()
    plot.tight_layout()
    plot.set_size_inches(6,5)
    plot.savefig(outfp, format='pdf')
    
for i in df.columns[1:7]:
    megahit_comp_plot(data=df, 
                      param=i, 
                      outfp=os.path.join(path, "%s.pdf" %i))
```
